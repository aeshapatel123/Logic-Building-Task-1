Problem 1: Unique Words in a Sentence

```python
print("Enter a sentence :")
sentence=input()          #taking sentence as input
words=sentence.split()    #Splitting the sentence into words
unique_words=set(words)   #Creating a set of unique words
print("Unique words count are: ",len(unique_words))  #printing the length of unique words
print("The unique words are: ",unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
employees_data={"Ravi":75000,"Anita":68000,"Kiran":72000,"Sahil":84000,"Aesha":69800}
for name,salary in employees_data.items():
    if salary==max(employees_data.values()):  #To calculate the maximum salary
        print("Highest Salary :",name,"-",salary )
```

Problem 3: Find Maximum and Minimum Values

```python
nums=[45,79,100,-5,0,34,12,90]  #list of numbers
minimum_val=nums[0]
maximum_val=nums[0]
for n in nums:
  if n<minimum_val:             #checking for minimum number
    minimum_val=n
  if n>maximum_val:             #checking for maximum number
    maximum_val=n
print("List:",nums)
print("Maximum Value: ",maximum_val)
print("Minimum Value: ",minimum_val)
```

Problem 4: Count Products Above a Price Threshold

```python
prices=[450,1200,899,1500,300,4999,1000]
count=0
for i in prices:
  if(i>1000):    #checking for product price above 1000
    count+=1
print("Products above price 1000 is :",count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance=['P','P','A','P','P']
total_days=len(attendance)   #calculating number if days
total_present=0
for i in attendance:
  if i=='P':               #checking for present days
    total_present+=1
attendance_percentage=(total_present/total_days)*100  #calculating attendance percentage
print("Attendance Percentage: ",attendance_percentage)
```

Problem 7: Count Character Frequency

```python
print("Enter a word :")
word=input()      #entering a word
dict={}           #creating a empty dictionary
for i in word:    #for loop to loop through each character
  if i in dict:
    dict[i]+=1    #calculating the frequency of each character
  else:
    dict[i]=1
print(dict)      #printing the dictionary
```

Problem 10: Calculate Average Marks

```python
stud_marks=[56,79,80,76,89,59]      #list of student marks
sum_marks=sum(stud_marks)           #sum of marks
average_marks=sum_marks/len(stud_marks)  #calculating the average
round_val=round(average_marks,2)         #rounding of to two decimal
print("Average Marks: ",round_val)
```
