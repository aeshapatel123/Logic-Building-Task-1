Problem Statement 1: Employee Performance Bonus Eligibility


```python
employee={
    "Ravi":92,
    "Anita":88,
    "Kiran":92,
    "Suresh":85
}
max_emp_score=max(employee.values()) # we get the max employee score from this
top_name=[name for name,score in employee.items() if score==max_emp_score] # we get the name of the top performing employee(s)
names=", ".join(top_name)# joins the names with a comma if there are more that one name entries in the list, this is done for better formatting
print(f"Top Performers Eligible for Bonus: {names} (Score: {max_emp_score})") # printing the top performing employee and their score
```

Problem Statement 2: Search Query Keyword Analysis

```python
import string
print("Enter your query")
query=input() #taking the query input
lower_query=query.lower() #converting to the lower case
cleaned=lower_query.translate(str.maketrans('','',string.punctuation)) #removeing punctuation
words=cleaned.split()
freq={}
for word in words:
    if word in freq:
        freq[word]+=1          #counting the frequency
    else:
        freq[word]=1
res={}
for word,count in freq.items():
    if count>1:
        res[word]=freq[word]
print(res)
```

Problem Statement 3: Sensor Data Validation

```python
sensor_reading=[3,4,7,8,10,12,5]
valid_readings=[]
n=len(sensor_reading) #calculating length of the readings
for i in range(n):
  if sensor_reading[i]%2==0: #calculating for even readings
    valid_readings.append((i,sensor_reading[i]))
print("Valid Sensor Readings (Hour,Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails=["ravi@gmail.com","sahil@gmail.com","aesha@yahoo.com","meena@yahoo.com","meena@gmail.com"]
domain_count={}
for e in emails:
  domain=e.split("@")[1]   #splitting domain name

  if domain in domain_count:
    domain_count[domain]+=1 #checking for domain name presence
  else:
    domain_count[domain]=1
total_emails=len(emails)
for domain in domain_count:
  per=(domain_count[domain]/total_emails)*100   #calculating the percentage of each domain name
  print(f"{domain}: {int(per)}%")
```

Problem Statement 5: Sales Spike Detection

```python
sales=[1200,1500,900,2200,1400,3000]
total_days=len(sales)
avg_sales=sum(sales)/total_days
threshold=avg_sales*1.30 #calculating 30% of the average
for i in range(total_days):
  if sales[i]>threshold:
    print(f"Day{i+1} : {sales[i]}")
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids=["user1","user2","user1","user3","user1","user3"]
count_ids={}
for uid in user_ids:
  if uid in count_ids:     #checking for the presence of the user ID
    count_ids[uid]+=1
  else:
    count_ids[uid]=1
for uid in count_ids:
  if count_ids[uid]>1:
    print(uid,"->",count_ids[uid], "times")
```
