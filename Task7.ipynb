Problem 1: Smart Light Controller

```python
class Control: # create a class for smart light control
  def __init__(self, light_name):
    self.light_name = light_name  #
    self.status="off"             # Light status is OFF by default

  def control_light(self,action): # Function to know about the light status
    if action=="on":
      self.status="ON"
    else:
      self.status="OFF"

  def display_status(self):      # Function to display the status
    print(self.light_name + " is "+ self.status)

light_name=input("Light Name: ") # Input the light name
action=input("Action (ON/OFF): ").lower() # Input the action perfromed on the light
light=Control(light_name) # Object 'light' created for the Control class
light.control_light(action) # Calling Method
light.display_status() # Calling Method
```

Problem 2: Employee ID Card System

```python
class ID_system:  # Class to display the employee id details
  def __init__(self,emp_name,emp_id,dept_name): # Constructor to store details
    self.emp_name=emp_name
    self.emp_id=emp_id
    self.dept_name=dept_name

  def display_details(self): # Function to display the id card details
    print()
    print("Employee ID Card")
    print("Name: ",self.emp_name)
    print("ID: ",self.emp_id)
    print("Department : ",self.dept_name)

emp_name=input("Employee Name: ") # Input employee name
emp_id=input("Employee ID: ")     # Input employee id
dept_name=input("Department Name: ") # Input department name
emp_obj=ID_system(emp_name,emp_id,dept_name) # creating object
emp_obj.display_details() # Calling Method

```

Problem 3: Mobile Contact Record

```python
class phone_record: # Create a class for phone record system
  def __init__(self,person_name,phone_number):
    self.person_name=person_name
    self.phone_number=phone_number

  def display_details(self): # Method to display the contact details for the person
    print()
    print("Contact Saved")
    print("Name: ",self.person_name)
    print("Phone Number: ",self.phone_number)

name=input("Contact Name:") # Input the contact name
phone=int(input("Phone Number:")) # Input the phone number
obj=phone_record(name,phone) # Object creation
obj.display_details() # Method calling
```

Problem 4: Product Price Tag Generator

```python
class product_price: # Creating class
  def __init__(self,product_name,product_price):
    self.product_name=product_name
    self.product_price=product_price

  def display_product_details(self): # Function to display product details
    print()
    print("Product:",self.product_name)
    print("Price: ₹",self.product_price)

name=input("Product Name: ") # Input Product name
price=int(input("Price: ")) # Input Product price
obj=product_price(name,price) # Object created
obj.display_product_details() # Method call
```

Problem 5: Movie Rating Display System

```python
class movie: # Create class
  def __init__(self,movie_name,rating):
    self.movie_name=movie_name
    self.rating=rating

  def display_movie(self): # Method to display the movie name and the rating of the movie
    print()
    print("Movie:",self.movie_name)
    print("Rating: ",self.rating)

name=input("Movie Name: ") # Input the movie name
rating=float(input("Rating: ")) # Input the rating as float
obj=movie(name,rating) # Object Creation
obj.display_movie() # Method call
```

Problem 6: Delivery Address Manager

```python
class address_manager:
  def __init__(self,address,customer_name):
    self.address=address
    self.customer_name=customer_name

  def display_address(self): # Method for displaying the customer address details
    print()
    print("Customer Delivery Details ")
    print("Name:",self.customer_name)
    print("Address:",self.address)

name=input("Customer Name: ") # Input customer name
address=input("Address: ")    # Input Address
obj=address_manager(address,name) # Object Creation
obj.display_address() # Method Calling
```
