Problem Statement 1:Smart Parking Lot Management System

```python
def parking_system(total_capacity,log_status): # function to make a system for a parking log system based on logs
  total_in=0      # To store the total parked vehicles at a current time
  peak=0
  for i in log_status:               # To loop through each input in the logs
    if i=="IN":
      total_in+=1                    # Increment the total in vehicle by 1
    elif i=="OUT" and total_in>0:
      total_in-=1                    # Decrement the total in vehicle by 1, when a car log is OUT

    if total_in>peak:   # Check for peak parkig usage
      peak=total_in

  print("Peak Parking Usage:",peak) # Display of peak parking usage
  print("Currently Parked Vehicles:",total_in) # Display of currently parked vehicles
  if total_in>total_capacity:
    print("Parking Status: Unvailable") # Printing the status of the parking lot
  else:
    print("Parking Status: Available")

print("Enter the parking capacity") # Entering the total capacity of the parking lot
total_capacity=int(input())
print("Enter the Vehicle Logs(IN/OUT) Space Seperated") # Entering the vehicle log
log_status=input().split()
parking_system(total_capacity,log_status) # Calling the function
```

Problem Statement 2:Online Food Delivery Time Estimator

```python
def Online_food_delivery(distance,traffic,weather): # Function to calculate ETA based on the above formed conditions
  min_prep_time=20 # Base meal preparation time needed to the restaurant
  base_time=(distance/25)*60 + (min_prep_time) # Base time calculated based on the distance, average speed = 25km/hr and the meal prep time
  # Calculating the extra time needed based on the traffic status
  if(traffic=="low"):
    base_time+=5
  elif(traffic=="medium"):
    base_time+=10
  elif(traffic=="high"):
    base_time+=15

  # Calculating the extra time needed based on the weather conditions
  if(weather=="clear"):
    base_time+=0
  elif(weather=="rainy"):
    base_time+=10
  elif(weather=="storm"):
    base_time+=20

  # Printing the estimated delivery time
  print(f"Estimated Delivery Time: {base_time} Minutes")

print("Distance(km):")
distance=float(input()) # Input for distance
print("Traffic Level(Low/Medium/High):")
traffic=input().lower() # Input the traffic status
print("Weather Condition(Clear/Rainy/Storm):")
weather=input().lower() # Input the weather status
Online_food_delivery(distance,traffic,weather)
```

Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor():  # Function to monitor resource usage
  resources={} # Empty dictionary to store resource usage
  n=int(input("Enter the number of resources:")) # Entering the total number of resources
  for i in range(n): # Loop to enter resource name and hours used and store it in the dictionary
    name=input("Enter the name of the resource:")
    usage=int(input("Enter the usage hours:"))
    resources[name]=usage
  overused=[] # List to store overused resources
  for name,usage in resources.items(): # Check each resource for usage
    if usage>8:
      overused.append(name)
  if overused:
    print("Overused Resources:",",".join(overused)) # Display results
    print("Energy Alert: Yes")
  else:
    print("Overused Resources: None")
    print("Energy Alert: No")
classroom_resource_monitor()   # Function Call
```

Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_registration_capacity_controller(): # Function for event registration control
  capacity=int(input("Enter Event Capacity:")) # Entering the total event capacity
  registrations=int(input("Enter Number of Registrations:")) # Enter the total registrations
  confirmed=0
  waitlist=0
  for i in range(registrations): # Loop  through each registration
    if confirmed<capacity:
      confirmed +=1
    else:
      waitlist+=1
  if confirmed==capacity:      # Determine registration status
    status="Closed"
  else:
    status="Open"
  print("Confirmed Registrations:",confirmed) # Display result
  print("Waitlist Users:",waitlist)
  print("Registration Status:",status)
event_registration_capacity_controller() # Function call
```
