Problem 1: Social Media – Post Engagement Analyzer


```python
def engagement(likes): #function to calculate post engagement based on likes
  sum_likes=0
  for like in likes:
    sum_likes+=like   #calculating the sum of the likes from different socail media platforms
  if sum_likes>=1000: #checking for post engagement based on the total number of likes
    return "Viral Post"
  else:
    return "Noraml Engagement"
print("Enter likes from different social media platforms")
likes=[] #list to store all likes from different social media plaforms (assuming 5 platfroms)
for i in range(5):
  like=int(input()) #taking the likes as input
  likes.append(like)
print("Total Likes: ",sum(likes)) #displaying the total likes
res=engagement(likes)
print("Post Status:" ,res) #displaying the post status
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rain_analyzer(rainfall_data,required_level): #function to check the rainfall status
  sum_rainfall=0
  for rain in rainfall_data:
    sum_rainfall+=rain   #calculating the sum of the rainfall from the list of the rainfall data
  avg_rainfall=sum_rainfall/len(rainfall_data) #calculating the average of the rainfall
  if avg_rainfall>=required_level: #checking for the rainfall status
    status="Adequate Rainfall"
  else:
    status="Inadequate Rainfall"
  print("Average Rainfall: ",avg_rainfall)
  print("Rainfall Status: ",status)
rainfall_data=[70,84,65,40,50]   #assumed list of rainfall of 5 days
required_level=72  #assumed required level of rainfall
rain_analyzer(rainfall_data,required_level)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def detect_username(usernames): #function to detect for duplicate usernames in a list of usernames
  unique_usernames=set(usernames) # we convert the list into a set because a set removes duplicate values and creates a unique list
  if len(usernames)!=len(unique_usernames): # we check the length of the original list and the set if unique values
    print("Duplicate Account Found : Yes")
  else:
    print("No Duplicate Account Detected")
users=["user1","user2","user3","user1","user4","user1"] #assumed list that contains the usernames
detect_username(users) #calling the defined function
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(prices): #function to filters crop prices above a premium threshold.
  premium_crops=[] #empty list to store premium crop prices
  for price in prices: #loop through each crop price in the list
    if price>2000:
      premium_crops.append(price)
  return premium_crops
crop_prices=[1500,2500,1800,3200,1700] #example list of all crop prices
premium_crops=filter_premium_crops(crop_prices) #calling function
print("Premium Crops:",premium_crops) #printing results
```

Problem 7: System Monitoring – Application Health Checker

```python
def health_checker(error_count): #function to check the system health based on the error counts
  if error_count==0: #checking if the error count is 0
    status="Healthy" #displaying the status of the system
  elif error_count<=5:
    status="Minor Issues"
  else:
    status="Critical Issues"
  return status
print("Enter the number of errors") #entering the amount of errors in the system
error_count=int(input())
print("Error Count:",error_count) #displaying the error count
print("System Status:",health_checker(error_count)) #displaying the system status
```

Problem 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_checker(attendace_list):# Calculate the total number of classes from the length of the attendance list
  total_classes=len(attendace_list)
  total_present=0
  for a in attendace_list:  # Loop through each entry in the attendance list
    if a=="P":              # Check if the attendance entry is 'P' (Present)
      total_present+=1
  percentage=(total_present/total_classes)*100
  # Determine eligibility based on the attendance percentage
  if percentage>=75:
    status="Eligible"
  else:
    status="Not Eligible"

  print("Attendance Percentage:",percentage) # Print the calculated attendance percentage
  print("Exam Eligibility:",status)
att_list=["P","P","A","P","P"]
attendance_checker(att_list)
```
