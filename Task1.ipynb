```python
marks=[45,78,90,33,60]
n=len(marks)
count_pass=0
count_fail=0
for i in range(n):
  if(marks[i]>=50):
    count_pass+=1
  else:
    count_fail+=1
print("Total number of PASS students are : ",count_pass)
print("Total Number of FAIL Students are : ",count_fail)
```
