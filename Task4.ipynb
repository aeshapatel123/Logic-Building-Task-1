Problem 1: Mobile Recharge Validation System

```python
def recharge_validation(recharge_value):
  if recharge_value<=50: #checks if amount is greater than 50 or not
    return "Invalid recharge amount : Amount must be greater than 50"
  if recharge_value in [199,299,399,599]: #checks if the amount entered is a valid plan or not
    return f"Recharge request suceesfull. recharge of {recharge_value} is done"
  return "Invalid request : Try again "
while True: #Loop will run untill we break it
  recharge_value=int(input("Enter the recharge amount : "))
  print(recharge_validation(recharge_value))

  if recharge_value in [199,299,399,599]: # if this condition is true then the loop breaks
    break
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def cab_fare(km,time):
  base_fare=50       #base fare initialization
  dist_fare=base_fare+(km*12) #calculating the fare based on the distance
  if(8<=time<=11)or(17<=time<=20): #checking for the peak hour
    dist_fare=dist_fare+(0.25*dist_fare) #adding extra 25% to the fare if it is a peak hour
    print("Peak hours charges added")
  return dist_fare #returning the total fare
while True:
  km=float(input("Enter the distance in km:")) #taking the distance as the input
  time=float(input("Enter the time in 24 hour format:")) #taking time as input
  total=cab_fare(km,time)
  print(f"Total fare is {total}") #printing the total fare
  retry=input("Do you want to recalculate the fair? (yes/no)")
  if retry=="no":
    print("Thank youu!! ")
    break
```

Problem 6: Password Strength Checker

```python
def password_strength(password):
  if len(password)<8: #checks for the length of the password
    return "Weak Password"
  if not any(ch.isdigit() for ch in password): #checks for the presence of a digit in the password
    return "Weak Password"
  if not any(ch in "@#$" for ch in password): #checks for the presence of any one special character in the password
    return "Weak Password"
  return "Strong Password"
password=input("Enter the password :") #taking the password as input
print(password_strength(password)) #printing the result

```
